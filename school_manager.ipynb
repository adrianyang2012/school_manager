```python
#@title setup(click here to create a school)
teacher_password = 'apples and bananas and oranges and blueberries'
list_of_students = []
list_of_Clubs = []
new_news = []
published_news = []
```

```python
#@title classes(edit this if there are bugs)
class Student:
  def __init__(self,name,id,grade):
    self.name = name
    self.grade = grade
    self.absences = 0
    self.unfinished_homework = 0
    self.id = id
    self.clubs_joined = []
  def report_absence(self):
    self.absences +=1
  def next_grade(self):
    self.grade +=1
  def give_homework(self):
    self.unfinished_homework +=1
  def finished_homework(self):
    if self.unfinished_homework - 1 > -1:
      self.unfinished_homework -=1
    else:
      print('you do extra homework')
  def full_report(self):
    print(f"\t name of student:{self.name}")
    print("______________________________________")
    print(f"\t student id: {self.__int__}")
    print("______________________________________")
    print(f"\t grade:{self.grade}")
    print("______________________________________")
    print(f"\t absences:{self.absences}")
    print("______________________________________")
    print(f"\t unfinished_homework:{self.unfinished_homework}")
    print("______________________________________")
    print(f"\t clubs joined:{len(self.clubs_joined)}")
  @property
  def __int__(self):
    return self.id
class Club:
  def __init__(self,name,creator,id,max_members = None):
    self.president = creator
    self.name = name
    self.id = id
    self.max_members = max_members
    self.president.clubs_joined.append((id,name))
    self.members = [self.president]
  def add_members(self,student):
    if self.max_members-1 >= len(self.members):
      print(f"{student.name} has joined the club")
      self.members.append(student)
      student.clubs_joined.append((self.id,self.name))
    else:
      print("sorry you can't join")
  def full_report(self):
    print(f'club name:{self.name}')
    print(f'creator : {self.president.name}')
    print('members:')
    for members in self.members:
      print(f'\tmember : {members.name}')
    print(f'id:{self.id}')
```
